The **Arrows2D** primitive produces polygons that can directly have their size modified.

```python
import numpy as np
%matplotlib inline
import plato
import plato.draw.matplotlib as draw
```

```python
thetas = np.linspace(0, 2*np.pi, 32)

xs = 4*np.cos(thetas)
ys = 4*np.sin(thetas)

positions = np.array([xs, ys]).T
colors = np.ones((len(thetas), 4))
colors[:, :3] = plato.cmap.cubeellipse(thetas, lam=.4)
magnitudes = .75 + .25*np.cos(5*thetas)

prim = draw.Arrows2D(positions=positions, colors=colors, 
                     angles=thetas, magnitudes=magnitudes)

scene = draw.Scene(prim, zoom=3)
scene.show()
```

The shape of the arrows created by the primitive can be changed by modifying its `vertices` attribute, which can also be used as an easy way to make the given positions correspond to the head or the tail of the arrows, rather than the center.

```python
positions = np.zeros((2, 2)) + .25
colors = np.ones((2, 4))*(.5, .5, .5, 1)
tails = draw.Arrows2D(positions=positions, angles=[0, np.pi/2])
heads = draw.Arrows2D.copy(tails)
heads.positions = positions - .25

tails.vertices = np.array(tails.vertices) + (.5, 0)
heads.vertices = np.array(heads.vertices) - (.5, 0)

scene = draw.Scene([heads, tails], zoom=10)
scene.show()
```
