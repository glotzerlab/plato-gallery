The **ConvexPolyhedra** primitive produces a collection of convex polyhedra with a common set of vertices.

```python
import itertools
import numpy as np
import plato
import plato.draw.povray as draw
import rowan

features = dict(antialiasing=.3)
```

```python
cube_vertices = list(itertools.product(*(3*[[-.5, .5]])))
oct_vertices = [np.roll((0, 0, v), i) for (i, v) in 
                itertools.product(range(3), [-1, 1])]

cubes = draw.ConvexPolyhedra(vertices=cube_vertices, 
                             positions=[(0, -.6, 0), (0, .65, 0)],
                             orientations=2*[(1, 0, 0, 0)],
                             colors=[(.5, .6, .7, 1), (.7, .6, .5, 1)])
octs = draw.ConvexPolyhedra(vertices=oct_vertices, positions=[(-1.5, .6, 0)])

rotation = rowan.from_axis_angle((.25, 1, 0), .5)
scene = draw.Scene([cubes, octs], zoom=7, 
                   rotation=rotation, features=features)
scene.show()
```

`ConvexPolyhedra` in many backends support an outline attribute:

```python
for prim in scene:
    prim.outline = .025
    
scene.show()
```
