These results are from the paper [Tunable emergent structures and traveling waves in mixtures of passive and contact-triggered-active particles](https://dx.doi.org/10.1039/C7SM00888K) by Agrawal, Bruss, and Glotzer. In this image, contact-triggered active particles (red) and passive particles (blue) assemble two different ordered structures.

```python
import gtar
import numpy as np

import plato
import plato.draw.matplotlib as draw
```

```python
with gtar.GTAR('CAPs.zip', 'r') as traj:
    positions = traj.readPath('position.f32.ind')
    box = traj.readPath('box.f32.ind')
    types = traj.readPath('type.u32.ind')
```

```python
colors = np.ones((len(positions), 4))
colors[types == 0, :3] = np.array([99., 138., 195.])/255
colors[types == 1, :3] = np.array([180., 80., 89.])/255
diameters = np.ones((len(positions),))
prim = draw.Disks(positions=positions[:, :2], colors=colors, diameters=diameters, outline=.1)

translation = (41.197945, -49.179943, 0)
scene = draw.Scene(prim, features=dict(pan=True, antialiasing=True), 
                   translation=translation, zoom=1)
target = '../gallery/contact_active_particles_matplotlib.png'
scene.save(target)
```
